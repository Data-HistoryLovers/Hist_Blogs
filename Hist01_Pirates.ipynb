# Pulling Tables from Wikipedia and Exploring the Tables

```python
import pandas as pd
import numpy as np

link = "https://en.wikipedia.org/wiki/List_of_pirates"
tables = pd.read_html(link)
len(tables)
toi = tables[2]
```

# Merging multiple list into dataframe and creating a new category for time periods

```python
frames = [tables[1], tables[2], tables[3],tables[4], tables[5], tables[6], tables[7], tables[8], tables[9],tables[10]]
result = pd.concat(frames)
result=result.reset_index()
roi = result.loc[result[4] == 'Comments']
roi=roi.reset_index()
roi2 = pd.merge(roi, per3, left_index=True, right_index=True)
result=result.reset_index()
results2 = pd.merge(result, roi2, how = 'left', left_on = result['level_0'], right_on = roi2['level_0'])
results3 = results2[['key_0', 0, 1, '2_x', '3_x', '4_x', '0_y']].copy()
results3['0_y'] = results3['0_y'].fillna(method='ffill')
results3.columns = ['count', 'Name', 'Life', 'Years Active', 'Country of Origin', 'Comments', 'Age_per']
results3 = results3[results3["Comments"] != 'Comments']
results3
```

```python
##Life Start and End clean up
temp1 = results3['Life'].str.split('–', n = 1, expand = True)
temp1['L_end'] = np.where(temp1[0].str.contains('d.'), temp1[0], None)
temp1['L_end']= temp1['L_end'].replace('d.', '', regex=True)
temp1['L_end']= temp1['L_end'].replace(np.nan, 'Unknown', regex=True)
temp1.loc[temp1['L_end']== 'Unknown',"L_end"] = temp1[1]
temp1.loc[temp1['L_end'].str.contains('BC', na=False), "L_end"] = '-' + temp1['L_end']
#temp1['L_end']= temp1['L_end'].str.replace(r'[A-z?]', '', regex=True)
temp1['L_start'] = np.where(~temp1[0].str.contains("d. ", na = False), temp1[0], None)
temp1['L_start']= temp1['L_start'].replace('fl. |c.|early', '', regex=True)
temp1.loc[temp1['L_start'].str.contains('BC', na=False), "L_start"] = '-' + temp1['L_start'] 
temp1['L_start']= temp1['L_start'].replace("BC", '', regex=True)
##temp1['L_start']= temp1['L_start'].str.replace(r'[A-z?]', '', regex=True)
temp1.loc[temp1['L_end'].str.contains('-', na=False), "L_start"] = '-' + temp1['L_start']
temp1
```

```python
temp1['L_start_M'] = temp1['L_start'].apply(pd.to_numeric, errors='coerce')
temp1['L_end_M'] = temp1['L_end'].apply(pd.to_numeric, errors='coerce')
temp1['Life_Span'] = temp1['L_end_M'] - temp1['L_start_M']
temp1 = temp1.drop([0,1, 'L_start_M', 'L_end_M'], axis=1)
temp1
```

```python
##Years Active cleanup
temp2 = results3['Years Active'].str.split('–', n = 1, expand = True)
temp2['A_end'] = np.where(temp2[0].str.contains('d.'), temp2[0], None)
temp2['A_end']= temp2['A_end'].replace('d.', '', regex=True)
#temp2['A_end'] = np.where(temp2['A_end'].str.contains("century|Century", na = False), '', None)
temp2['A_end']= temp2['A_end'].replace(np.nan, 'Unknown', regex=True)
temp2.loc[temp2['A_end']== 'Unknown',"A_end"] = temp2[1]
temp2.loc[temp2['A_end'].str.contains('BC', na=False), "A_end"] = '-' + temp2['A_end']
#temp2['A_end']= temp2['A_end'].str.replace(r'[A-z?]', '', regex=True)
temp2['A_start'] = np.where(~temp2[0].str.contains("d. ", na = False), temp2[0], None)
#temp2['A_start'] = np.where(temp2['A_start'].str.contains("century|Century", na = False), '', None)
temp2['A_start']= temp2['A_start'].replace('fl. |c.', '', regex=True)
temp2.loc[temp2['A_start'].str.contains('BC', na=False), "A_start"] = '-' + temp2['A_start'] 
temp2['A_start']= temp2['A_start'].replace("BC", '', regex=True)
##temp2['A_start']= temp2['A_start'].str.replace(r'[A-z?]', '', regex=True)
temp2.loc[temp2['A_end'].str.contains('-', na=False), "A_start"] = '-' + temp2['A_start']
temp2
```

```python
temp2['A_start_M'] = temp2['A_start'].apply(pd.to_numeric, errors='coerce')
temp2['A_end_M'] = temp2['A_end'].apply(pd.to_numeric, errors='coerce')
temp2['Years_Active'] = temp2['A_end_M'] - temp2['A_start_M']
temp2 = temp2.drop([0,1,'A_start_M', 'A_end_M'], axis=1)
temp2
```

```python
results3 = pd.merge(results3, temp1, left_index=True, right_index=True)
results3 = pd.merge(results3, temp2, left_index=True, right_index=True)
results3
```

```python
results3['L_end_M'] = results3['L_end'].apply(pd.to_numeric, errors='coerce')
results3['A_end_M'] = results3['A_end'].apply(pd.to_numeric, errors='coerce')
results3['pir_d_cnt'] = results3['A_end_M'] - results3['L_end_M']
results3
```

```python
##Died on last year of activity
results3['Pir_Death'] = None
results4 = results3
results4.loc[results4['pir_d_cnt']== 0,"Pir_Death"] = 'Yes'
results4.loc[results4['pir_d_cnt']!= 0,"Pir_Death"] = 'No'
results4.drop(['L_end_M', 'A_end_M', 'pir_d_cnt'], axis = 1)
results4
```

```python
results4.to_csv('results.csv')
pir_d_a = results4.groupby(['Age_per','Pir_Death']).count().reset_index()
pir_d_a
```
